```python
# First let's import the packages we will use in this project


import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# Now we need to read in the data

df = pd.read_csv(r'C:\Users\josue\Documents\Portfolio Projects\Movies\movies.csv')
```

```python
df
```

```python
# Loop through the data and see if there is anything missing

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Delete rows with missing values

df.dropna(inplace=True)
```

```python
# Data Types for our columns

print(df.dtypes)
```

```python
# Change data type of columns

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
#Create correct year column

df['yearcorrect'] = df['released'].str.split(',').str[1].str[:4]
```

```python
# Are there any Outliers?

df.boxplot(column=['gross'])
```

```python
df.drop_duplicates()
```

```python
# Order our Data a little bit to see

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x="gross", y="budget", data=df, scatter_kws={"color": "red"}, line_kws={"color":"blue"})
```

```python
# Plot score vs gross using seaborn

sns.regplot(x="score", y="gross", data=df, scatter_kws={"color": "red"}, line_kws={"color":"blue"})
```

```python
# Create a new DataFrame excluding the first row
df_numeric = df.iloc[1:].select_dtypes(include='number').apply(pd.to_numeric, errors='coerce')

# Calculate the correlation matrix for the numeric columns
df_numeric.corr(method ='pearson')
```

```python
df_numeric.corr(method ='kendall')
```

```python
df_numeric.corr(method ='spearman')
```

```python
correlation_matrix = df_numeric.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
# Using factorize - this assigns a random numeric value for each unique categorical value

df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)
```

```python
# We can now take a look at the ones that have a high correlation (> 0.5)

strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```
